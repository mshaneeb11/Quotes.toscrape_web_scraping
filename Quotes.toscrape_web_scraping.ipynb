```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage=requests.get("https://quotes.toscrape.com/").text
soup = BeautifulSoup(webpage,"lxml")
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.find_all("h1")
```

```python
quotes=[]
for i in soup.find_all("span",class_="text"):
    quotes.append(i.text)
```

```python
print(quotes)
```

```python
authors=[]
for i in soup.find_all("small",class_="author"):
    authors.append(i.text)
```

```python
print(authors)
```

```python
tags = []
for i in soup.find_all("a",class_="tag"):
    tags.append(i.text)
```

```python
print(tags)
```

```python
d1 = {
    "Quotes":quotes,
    "Authors":authors,
}
```

```python
d1
```

```python
df = pd.DataFrame(d1)
```
